## Agregado de reverberación a un archivo de audio

Veamos cómo podemos usar Pedalboard para agregar efecto de eco a nuestro audio de trompeta:

```python
from pedalboard import Pedalboard, Chorus, Reverb
from pedalboard.io import AudioFile

# Read in a whole audio file:
with AudioFile('./trumpet.wav', 'r') as f:
  audio = f.read(f.frames)
  samplerate = f.samplerate

# Make a Pedalboard object, containing multiple plugins:
board = Pedalboard([Chorus(), Reverb(room_size=0.55)])

# Run the audio through this pedalboard!
effected = board(audio, samplerate)

# Write the audio back as a wav file:
with AudioFile('./trumpet_reverb.wav', 'w', samplerate, effected.shape[0]) as f:
  f.write(effected)
```

Ahora escuchemos la versión original y la que tiene el efecto agregado:

```python
import IPython.display as ipd
print("Trompeta Original")
ipd.display(ipd.Audio(data=audio, rate=samplerate))

print("\n\n Trompeta con Reverberación")
ipd.display(ipd.Audio(data=effected, rate=samplerate))
```
