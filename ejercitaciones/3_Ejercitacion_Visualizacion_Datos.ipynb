## Ejercitación: Visualización por grupos

Algo muy útil al hacer análisis exploratorio de datos es realizar gráficos comparativos agrupando de acuerdo de alguna variable de interés.

Carguemos el dataset con las Notas que usamos la clase anterior.

```python
df = pd.read_excel('../datos/notas.xlsx')
df.head()
```

*Tareas:*

- Graficar mediante un histograma la cantidad de alumnos en cada condición
- Hacer un boxplot para conocer la distribución de las notas del 1º Parcial de acuerdo a la variable de agrupamiento Año (en el eje x)
- Agregar una nueva variable de agrupamiento (argumento hue) que sea el Curso
- Graficar un mapa de calor con las correlaciones de las variables: 1º Parcial, 2º Parcial, 1º Recuperatorio de 1º Parcial y 1º Recuperatorio de 2º Parcial

Hay otro de gráfica interesante que es conocida como **Violin Plot**. Muestra la distribución de los datos junto a un boxplot en su interior.
- Hacer un conjunto de violin plots (en subplots) donde se muestren las notas del 1º Parcial y del 2º Parcial agrupadas por Curso y Modalidad

### Soluciones

```python
sns.histplot(df, x='Condicion')
```

```python
sns.boxplot(df, x='Año', y='1P')
```

```python
sns.boxplot(df, x='Año', y='1P', hue='Curso', width=0.75)
```

```python
cols = ['1P', '2P', '1P_R1', '2P_R1']  # variables a correlacionar
corr = df[cols].corr(method='spearman')
sns.heatmap(np.abs(corr), annot=corr, cmap='Blues')
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.violinplot(df, x='Curso', y='1P', hue='Modalidad') 
plt.ylabel('Notas')
plt.title('1º Parcial')
plt.subplot(1,2,2)
sns.violinplot(df, x='Curso', y='2P', hue='Modalidad') 
plt.ylabel('Notas')
plt.title('2º Parcial')
```
